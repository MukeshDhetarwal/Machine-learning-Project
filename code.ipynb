```python
numerical_columns = ['Applicant_City_PIN','Manager_Grade','Manager_Num_Application','Manager_Num_Coded','Manager_Business','Manager_Num_Products','Manager_Business2','Manager_Num_Products2']
```

```python
from sklearn.impute import SimpleImputer
```

```python
train[numerical_columns] = imputer.fit_transform(train[numerical_columns])
test[numerical_columns] = imputer.fit_transform(test[numerical_columns])
```
